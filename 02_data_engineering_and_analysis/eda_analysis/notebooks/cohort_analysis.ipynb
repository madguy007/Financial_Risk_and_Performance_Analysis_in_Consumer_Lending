## cohort Analysis

### Categorization

Since the dataset only contains loans issued in January across years, monthly cohorting is not feasible. Therefore, I grouped users by year to maintain time-based cohort structure.

```python
df['cohort_year'] = df['issue_date'].dt.year
```

```python
df.head()
```

#### year wise total customers

```python
cohort_size = df.groupby('cohort_year')['id'].nunique()
```

```python
cohort_size
```

####loan lifecycle engagement retention

#### Since user activity logs were unavailable, I approximated engagement using repayment behaviour. Users who made payments and had a good loan status were considered active, as they are still interacting with the platform financially

👉 Cohort analysis needs consistent behavior tracking

And repayment is:

- Measurable ✅
- Available ✅
- Time-related ✅

```python
df['is_active'] = (df['total_pymnt'] > 0) & (df['loan_condition'] == 'GOOD LOAN')
```

```python
active_users = df[df['is_active']].groupby('cohort_year')['id'].nunique()
```

```python
active_users
```

#### Summarization

```python
cohort_df = pd.concat([cohort_size, active_users], axis=1)
cohort_df.columns = ['total_users', 'active_users']
```

```python
cohort_df['retention_rate'] = cohort_df['active_users'] / cohort_df['total_users']
```

## Question 2 : If retention drops after first month, why + how to fix?

# 🧠 Step 1: Understand in YOUR context

In app world:
→ Users stop using app after signup

In YOUR dataset:
→ Users stop repaying / become bad loans

## 🎯 Translation

Retention drop =
→ Users become inactive (no payment / bad loan)


# ✅ STEP 1: Identify Possible Reasons

## 🔴 1. User Quality Issue (Most Important)
- High DTI (dti)
- Low grade (grade)

Meaning:
→ Risky borrowers  
→ Cannot sustain repayment  





High DTI check

```python
df['dti_bucket'] = pd.cut(df['dti'], bins=[0,10,20,30,100],
                         labels=['Low','Medium','High','Very High'])
```

```python
dti_retention = df.groupby('dti_bucket').agg(
    total_users=('id','nunique'),
    active_users=('is_active','sum')
)

dti_retention['retention_rate'] = dti_retention['active_users'] / dti_retention['total_users']
```

```python
dti_retention
```

Low Grade check

```python
grade_retention = df.groupby('grade').agg(
    total_users=('id','nunique'),
    active_users=('is_active','sum')
)

grade_retention['retention_rate'] = grade_retention['active_users'] / grade_retention['total_users']
```

```python
grade_retention
```


## 🔴 2. Pricing Issue
- High interest_rate

Meaning:
→ EMI too high  
→ Users drop early  


## 🔴 3. Product / Experience Issue
- Poor onboarding
- Confusion about repayment
- No reminders




## 🔴 4. Wrong Targeting
- Low income users
- Unstable employment (emp_length_int)




```python
income_retention = df.groupby('income_category').agg(
    total_users=('id','nunique'),
    active_users=('is_active','sum')
)

income_retention['retention_rate'] = income_retention['active_users'] / income_retention['total_users']
```
