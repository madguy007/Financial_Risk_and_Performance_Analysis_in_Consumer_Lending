## C. Load Preparation Phase (Data Structuring & Export Readiness)

1. Add a default_rate_indicator column that computes the ratio of defaulted loans (Bad Loan) to total loans within the same year.

```python
# Add new column
cursor.execute("""
ALTER TABLE loans
ADD COLUMN default_rate_indicator REAL;
""")
conn.commit()
```

```python
# Calculating default rate.
cursor.execute("""
UPDATE loans
SET default_rate_indicator = (
    SELECT default_rate
    FROM (
        SELECT year,
               SUM(risk_flag)*1.0/COUNT(*) AS default_rate
        FROM loans
        GROUP BY year
    ) t
    WHERE t.year = loans.year
);
""")
conn.commit()
```

```python
pd.read_sql('select default_rate_indicator from loans limit 5', conn)
```

2. Extract loan term as numeric value (e.g., convert ‘36 months’ → 36).

```python
cursor.execute("""
ALTER TABLE loans
ADD COLUMN term_months INTEGER;
""")
conn.commit()
```

```python
cursor.execute("""
UPDATE loans
SET term_months = CAST(TRIM(REPLACE(term, 'months', '')) AS INTEGER);
""")
conn.commit()
```

```python
pd.read_sql("""
SELECT term, term_months FROM loans LIMIT 5;
""", conn)
```

3. Add a new column income_to_loan_ratio calculated as the borrower’s annual income divided by the loan amount.

```python
# Add new column
cursor.execute("""
ALTER TABLE loans
ADD COLUMN income_to_loan_ratio REAL;
""")
conn.commit()
```

```python
# add the ratio calculation
cursor.execute("""
UPDATE loans
SET income_to_loan_ratio =
            CASE WHEN loan_amount > 0 THEN annual_inc * 1.0 / loan_amount
                 ELSE NULL
            END;
""")
conn.commit()
```

```python
pd.read_sql('select annual_inc, loan_amount,income_to_loan_ratio from loans limit 5', conn)
```

4. Create a new table loans_cleaned containing only cleaned and transformed records (no nulls in key fields).

```python
# check and drop if loans cleaned folder is there
cursor.execute("""
DROP TABLE IF EXISTS loans_cleaned;
""")

# Create new loans_cleaned folder with given columns form loans
cursor.execute("""
CREATE TABLE loans_cleaned AS
SELECT
    id,
    year,
    issue_date,
    final_date,
    term_months,
    emp_length_int,
    home_ownership,
    annual_inc,
    loan_amount,
    income_to_loan_ratio,
    purpose,
    interest_rate,
    interest_payments,
    grade,
    dti,
    total_pymnt,
    total_rec_prncp,
    recoveries,
    installment,
    region,
    profitability,
    loan_condition,
    risk_flag,
    default_rate_indicator
FROM loans
WHERE
    loan_amount IS NOT NULL
    AND annual_inc IS NOT NULL
    AND interest_rate IS NOT NULL
    AND purpose IS NOT NULL;
""")
conn.commit()
```

```python
pd.read_sql('select * from loans_cleaned limit 5', conn)
```
